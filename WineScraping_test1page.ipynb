```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
winepage = "https://www.winemag.com/?s=&drink_type=wine&varietal=Shiraz/Syrah&country=US&page=1&sort_by=pub_date_web&sort_dir=desc"
```

```python
hdr = {'User-Agent': 'Mozilla/5.0'} #Not sure what this header does
req = Request(winepage,headers=hdr) #checks for User-Agent to be of "popular browser"
page = urlopen(req)
soup = BeautifulSoup(page, 'lxml')
print(soup)
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup.prettify())
```

```python
mydivs = soup.findAll("div", {"class": "title"})
```

```python
T = []
Y = []
for div in mydivs:
    T.append(div.text)
    x = re.sub("\D", "", div.text)
    x = int(x)
    Y.append(x)
```

```python
Y
```

```python
ratings = soup.findAll("span", {"class": "rating"})
```

```python
ratings
```

```python
R = []
for rating in ratings:
    x = re.sub("\D", "", rating.text)
    x = int(x)
    R.append(x)
```

```python
R
```

```python
price = soup.findAll("span", {"class": "price"})
```

```python
P = []
for entry in price:
    x = re.sub("\D", "", entry.text)
    x = int(x)
    P.append(x)
```

```python
P
```
