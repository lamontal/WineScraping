```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading necessary data files
sono= pd.read_csv("Sonoma_weather_month.csv")
coast= pd.read_csv("CCoast_weather_month.csv")
valley= pd.read_csv("CValley_weather_month.csv")
```

```python
sono2 = sono.groupby("date").mean() #some months have multiple stations recording data, decided to average all stations
coast2 = coast.groupby("date").mean()
valley2  = valley.groupby("date").mean()
```

```python
sono2["dt"] = pd.to_datetime(sono2.index)
sono2["Year"]=sono2["dt"].dt.year
sono2["Month"] = sono2["dt"].dt.month
```

```python
sono2
```

```python
sonoyear = sono2.groupby("Year")
```

```python
yeardic = {}
for group, item in sonoyear:
    monthdic = {}
    for index, row in item.iterrows():
        monthdic.update({"PRCP {}".format(row["Month"]): row["PRCP"] , 
                "TAVG {}" .format(row["Month"]) : row["TAVG"],
                "TMIN {}" . format(row["Month"]) : row["TMIN"],
                "TMAX {}" . format(row["Month"]) : row["TMAX"]})
    x= {group : monthdic}
    yeardic.update(x)
```

```python
yeardf = pd.DataFrame.from_dict(yeardic, orient='index') 
```
