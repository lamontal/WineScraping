```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading necessary data files
sono= pd.read_csv("Sonoma_weather_month.csv")
coast= pd.read_csv("CCoast_weather_month.csv")
valley= pd.read_csv("CValley_weather_month.csv")

locations = [sono, coast, valley]
```

```python
cleandfs= []
for location in locations:

    df = location.groupby("date").mean() #some months have multiple stations recording data, decided to average all stations

    #Making index a datetime and adding columns of year and month
    df["dt"] = pd.to_datetime(df.index)
    df["Year"]=df["dt"].dt.year
    df["Month"] = df["dt"].dt.month
    
    
    dfyear = df.groupby("Year") #grouping by year to prep for new dataframe with years as indices
    yeardic = {}
    for group, item in dfyear:
        monthdic = {}
        for index, row in item.iterrows():
            monthdic.update({"PRCP {}".format(row["Month"]): row["PRCP"] , 
                "TAVG {}" .format(row["Month"]) : row["TAVG"],
                "TMIN {}" . format(row["Month"]) : row["TMIN"],
                "TMAX {}" . format(row["Month"]) : row["TMAX"]})
        x= {group : monthdic}
        yeardic.update(x)
    yeardf = pd.DataFrame.from_dict(yeardic, orient='index') 
    cleandfs.append(yeardf)
```
